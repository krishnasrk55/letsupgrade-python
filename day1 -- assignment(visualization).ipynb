# Question 2

```python
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2

plt.plot(days,sales_1)
plt.title("Days vs Sales_1")
plt.xlabel("Days")
plt.ylabel("Sales_1")
```

```python
plt.plot(days,sales_2)
plt.title("Days vs Sales_2")
plt.xlabel("Days")
plt.ylabel("Sales_2")
```

```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]
```
